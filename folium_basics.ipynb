```python
import numpy as np
import pandas as pd
import folium
import requests
import io
```

```python
# Import folium MarkerCluster plugin
from folium.plugins import MarkerCluster
# Import folium MousePosition plugin
from folium.plugins import MousePosition
# Import folium DivIcon plugin
from folium.features import DivIcon
```

## Load dataset

```python
# Download and read the `spacex_launch_geo.csv`
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'

response = requests.get(URL)
csv_data = io.StringIO(response.text)
spacex_df = pd.read_csv(csv_data)
spacex_df.head()
```

```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

## Mark locations in cluster form

```python
marker_cluster = MarkerCluster()
```

```python
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
def add_marker_color(spacex_df):
    spacex_df['marker_color'] = np.where(spacex_df['class'] == 1, 'green', 'red')
    return spacex_df

spacex_df = add_marker_color(spacex_df)
spacex_df.tail(10)
```

```python
# Initialize the map
site_map = folium.Map()
```

```python
for index, record in spacex_df.iterrows():
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        icon=folium.Icon(color='white', icon_color=record['marker_color'])
    )
    # Aadir el marcador al cluster de marcadores
    marker_cluster.add_child(marker)

site_map.add_child(marker_cluster)
site_map
```
